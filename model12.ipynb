```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')

# Load data
df = pd.read_csv("CreditCardUsageP.csv")

# Check new data
print(df.columns)
print(df.head())

# Plot the relationship
plt.figure(figsize=(20, 5))
plt.subplot(1, 3, 1)
sns.boxplot(x=df['PRC_FULL_PAYMENT'], y=df['BALANCE'])
plt.title('Balance Distribution by Full Payment Rate')
plt.show()

# Encode categorical data
def encode_labels(data):
    le = LabelEncoder()
    for col in data.columns:
        if data[col].dtype == 'object':
            data[col] = le.fit_transform(data[col])
    return data

df = encode_labels(df)

# Check and remove missing values
print(df.isna().sum())

# Remove null values
df.dropna(inplace=True)

# Determine the number of clusters 
inertia_t = []
for i in range(1, 12):
    km = KMeans(n_clusters=i)
    km.fit(df)
    inertia_t.append(km.inertia_)

# Plot to check the suggested number of clusters
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 2)
sns.lineplot(x=range(1, 12), y=inertia_t)
plt.title('KMeans Inertia on Transformed Data')
plt.show()

# Build the model
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, n_init=10, max_iter=500)
kmeans.fit(df)

# Add the clusters into the initial dataframe
df['clusters'] = kmeans.labels_

# Evaluate the model performance
sil_score = silhouette_score(df, kmeans.labels_)
print(f'Silhouette Score: {sil_score}')

# Plot the clusters/results based on two selected features
plt.figure(figsize=(20, 5))
plt.scatter(df["BALANCE"], df["PURCHASES"], c=kmeans.labels_, cmap="rainbow")
plt.xlabel("Balance")
plt.ylabel("Purchases")
plt.title("Cluster Distribution by Balance and Purchases")
plt.show()

# Another plot based on different features
plt.figure(figsize=(10, 5))
plt.scatter(df["CREDIT_LIMIT"], df["PRC_FULL_PAYMENT"], c=kmeans.labels_, cmap="rainbow")
plt.xlabel("Credit Limit")
plt.ylabel("Full Payment Rate")
plt.title("Cluster Distribution by Credit Limit and Full Payment Rate")
plt.show()
```
